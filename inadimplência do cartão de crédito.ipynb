# Prever a inadimplência do cartão de crédito.

Nesse caso, estamos tentando **prever** se uma pessoa provavelmente entrará em default (classe = 1) ou não (classe = 0).

### Algumas importações necessárias.

São poucas bibliotecas utilizadas aqui. É apenas biblioteca de trabalhar com dataframe e biblioteca que designa o aprendizado de máquinas.

```python
import pandas as pd
import geopandas as gpd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

### E os dados?

Os dados são fornecidos pelo Kaggle (https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud).

```python
#importando o dataframe para a váriavel arquivo.
arquivo = pd.read_csv("C:\\Users\\engce\\OneDrive\\Documentos\\ESTOQUE\\creditokaggle.zip").sample(10_000, random_state=390)
```

```python
arquivo
```

## A que sentido deve ser levado?

Para construção do modelo, é preciso especificar que aspecto da o sentido do modelo. No caso, a inadimplência. A inadimplência é o que estamos avaliando. Para avaliar a indadimplência, existe a coluna classe, nela o modelo vai se basear, entendendo todo o dataframe e usando a coluna classe como base do modelo.

```python
#Aqui foi usado o train_test_split da biblíoteca sklearn.
train_df, test_df = train_test_split(arquivo, test_size=.2, random_state=413)

X_train = train_df.drop(columns=['Class'])
y_train = train_df['Class']
X_test = test_df.drop(columns=['Class'])
y_test = test_df['Class']
```

## Finalmente o Modelo?

O modelo de aprendizagem de máquina já está aqui. **Usando XGBClassifier**. 

```python
#Aqui é dado o modelo de predição.
model = XGBClassifier(use_label_encoder=False)
model.fit(X_train, y_train);
```

### O modelo já está pronto, acabou mesmo?

A partir daqui, é apenas usar o modelo de aprendizagem. Para início, é interessante acrescentar as colunas que apresentam as predições do modelo, ao dataframe do prórprio modelo.

```python
#Acrescentando as colunas.
pred_df = pd.DataFrame({
    "predicted_price": model.predict(X_test),
    "actual_price": y_test
})
pred_df.join(X_test)
```

# Conclusão.

### O modelo de aprendizagem de máquina foi feito. 

Isso abre portas para estudar casos a partir das predições encontradas. 

A qualidade do modelo, a previsão de indadiplencia é concluída.

```python
model.score(X_test, y_test)
```
